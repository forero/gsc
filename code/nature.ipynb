```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = np.loadtxt('../data/nature_index_global.csv',
                  dtype=[('country','|S20'),('AC', 'f8'),('FC', 'f8'), ('WFC', 'f8')])
```

```python
n_total = len(data)
ranking = np.arange(n_total)+1
plt.plot(ranking, data['WFC'])

comparison_id = [0,9,19,29,39]
for ii in comparison_id:
    plt.scatter(ranking[ii], data['WFC'][ii], s=100, label=data['country'][ii].decode())
    
c_id = data['country']==b'Colombia'
plt.scatter(ranking[c_id], data['WFC'][c_id], s=100, label='Colombia')
plt.legend(loc=3)
plt.semilogy()
plt.semilogx()
```
